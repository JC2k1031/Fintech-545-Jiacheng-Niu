```python
import numpy as np
from statsmodels.tsa.arima_process import ArmaProcess
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Simulate AR(1)
ar1 = ArmaProcess(ar=[1], ma=[1])
data1 = ar1.generate_sample(nsample=1000)

# Simulate AR(2)
ar2 = ArmaProcess(ar=[1, -0.2], ma=[1])
data2 = ar2.generate_sample(nsample=1000)

# Simulate AR(3)
ar3 = ArmaProcess(ar=[1, -0.2, 0.3], ma=[1])
data3 = ar3.generate_sample(nsample=1000)

# Simulate MA(1)
ma1 = ArmaProcess(ar=[1], ma=[0.1, 0.5])
data4 = ma1.generate_sample(nsample=1000)

# Simulate MA(2)
ma2 = ArmaProcess(ar=[1], ma=[0.1, 0.5, 0.3])
data5 = ma2.generate_sample(nsample=1000)

# Simulate MA(3)
ma3 = ArmaProcess(ar=[1], ma=[0.1, 0.5, 0.3, 0.2])
data6 = ma3.generate_sample(nsample=1000)

# Plot ACF and PACF for AR(1)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,6))
plot_acf(data1, lags=20, ax=ax1)
plot_pacf(data1, lags=20, ax=ax2)
plt.show()

# Plot ACF and PACF for AR(2)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,6))
plot_acf(data2, lags=20, ax=ax1)
plot_pacf(data2, lags=20, ax=ax2)
plt.show()

# Plot ACF and PACF for AR(3)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,6))
plot_acf(data3, lags=20, ax=ax1)
plot_pacf(data3, lags=20, ax=ax2)
plt.show()

# Plot ACF and PACF for MA(1)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,6))
plot_acf(data4, lags=20, ax=ax1)
plot_pacf(data4, lags=20, ax=ax2)
plt.show()

# Plot ACF and PACF for MA(2)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,6))
plot_acf(data5, lags=20, ax=ax1)
plot_pacf(data5, lags=20, ax=ax2)
plt.show()

# Plot ACF and PACF for MA(3)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,6))
plot_acf(data6, lags=20, ax=ax1)
plot_pacf(data6, lags=20, ax=ax2)
plt.show()
```
